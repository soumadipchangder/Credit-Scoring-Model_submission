```python
import numpy as np 
import pandas as  pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(['ID','Customer_ID','Month','Name','SSN','Monthly_Inhand_Salary','Num_Bank_Accounts','Num_Credit_Card','Interest_Rate','Num_of_Loan','Type_of_Loan','Delay_from_due_date','Num_of_Delayed_Payment','Changed_Credit_Limit','Num_Credit_Inquiries','Credit_Mix','Credit_Utilization_Ratio','Credit_History_Age','Payment_of_Min_Amount','Amount_invested_monthly','Payment_Behaviour','Monthly_Balance','Age','Occupation'],axis=1,inplace=True)
```

```python
df
```

```python
df['Credit_Score'].replace({"Good":1,"Poor":0,"Standard":2},inplace = True)
```

```python
df
```

```python
df.info()
```

```python

for col in ['Annual_Income', 'Outstanding_Debt']:
    df[col] = pd.to_numeric(df[col].str.replace(r'[^0-9.]', '', regex=True), errors='coerce')

df = df.fillna(0).astype(int)  

print(df.dtypes) 
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop('Credit_Score', axis=1)

y = df['Credit_Score']

clf = RandomForestClassifier(n_estimators=100, random_state=42)

clf.fit(X, y)
```

```python
clf.score(X, y)
```

```python
import pickle

with open('credit_scoring_model.pkl', 'wb') as f:
    pickle.dump(clf, f)
    f.close()
```
